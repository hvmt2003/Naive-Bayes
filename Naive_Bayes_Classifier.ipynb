```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv('/content/placement_cgpa_score_placed.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='cgpa',y='resume_score',data=dataset,hue='placed')
plt.legend()
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#to check if normally distributed or not

sns.kdeplot(data=dataset['cgpa'])
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test), gnb.score(x_train,y_train)
```

```python
gnb.predict([[6.17,5.17]])
```

```python
plt.figure(figsize=(6,5))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
```

```python
mnb=MultinomialNB()
```

```python
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test), mnb.score(x_train,y_train)
```

```python
plt.figure(figsize=(6,5))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
```

```python
bnb=BernoulliNB()
```

```python
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test), bnb.score(x_train,y_train)
```

```python
plt.figure(figsize=(6,5))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
